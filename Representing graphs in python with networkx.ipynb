# Representing graphs in python with networkx

```python
#import networkx library  
import networkx as nx 

#https://networkx.org/documentation/stable/tutorial.html

#import matplotlib
import matplotlib.pyplot as plt 
```

####  Graph nodes can be : 
   - strings , images, XML objects , another graph , nodes objects ...etc

### Create an empty graph 

```python
G = nx.Graph()
```

### Add a node: 

```python
G.add_node('a')
```

### Add nodes from list 

```python
G.add_nodes_from(['b', 'c', 'd'])
```

### Print the graph nodes 

```python
print('Graph G nodes are : ')
print(G.nodes())
```

```python
print('Graph G edges : ')
print(G.edges())
```

### Graph can also be created by adding, link by link (edge by edge)

```python
G.add_edges_from([('a','c'),('a','b'),('b','c'),('b','e'),('b','d'),('c','d'),('c','e'),('d','e'),('d','f'),('e','f')])
```

### Print the graph nodes 

```python
print('Graph G nodes are : ')
print(G.nodes())

#print the graph edges 
print('Graph G edges : ')
print(G.edges())
```

### Draw the graph (with labels) 

```python
nx.draw(G, with_labels=True)
plt.show()
```

### Draw a cubical graph 

#### In the mathematical field of graph theory, a cubic graph is a graph in which all vertices have degree three. In other words, a cubic graph is a 3-regular graph. Cubic graphs are also called trivalent graphs. 

```python
G = nx.cubical_graph()
subax1 = plt.subplot(121)
nx.draw(G)
subax2 = plt.subplot(122)
nx.draw(G, pos = nx.circular_layout(G), node_color='r', edge_color='b')
```

### Remove a node from the graph 

```python
G = nx.Graph()
G.add_nodes_from(['a','b', 'c', 'd'])
G.add_edges_from([('a','c'),('a','b'),('b','c'),('b','e'),('b','d'),('c','d'),('c','e'),('d','e'),('d','f'),('e','f')])


#remove a node from the graph
G.remove_node('a')

nx.draw(G,with_labels = True)
```

### Remove an edge from the graph

```python
G.remove_edge('b','e') 

nx.draw(G, with_labels = True)
```

## Path graph : 

#### The path graph is a tree with two nodes of vertex degree 1, and the other. nodes of vertex degree 2. A path graph is therefore a graph that can be drawn so that all of its vertices and edges lie on a single straight line (Gross and Yellen 2006)

### Create a path graph

```python
Gpathgraph = nx.path_graph(5)

print('Nodes of the path graph : ')
print(Gpathgraph.nodes())

print('Edges of the path graph : ')
print(Gpathgraph.edges())

#draw the graph 
nx.draw(Gpathgraph,with_labels=True)
plt.show()
```

## Petersen Graph: 

#### Petersen graph is an undirected graph with 10 vertices and 15 edges

### Create a petersen graph 

```python
Gpetersengraph = nx.petersen_graph()

print('Nodes of the petersen graph:')
print(Gpetersengraph.nodes())
print('Edges of petersen graph:')
print(Gpetersengraph.edges())
nx.draw(Gpetersengraph,with_labels=True)
plt.show()
```

### Play with fonts and colors

```python
nx.draw_shell(Gpetersengraph,nlist=[range(6,11),range(6)], with_labels=True, font_weight='bold')
plt.show()
```

### with options , we can modify many style attributes

```python
options = {'edge_color': 'red', 'node_color': 'white', 'node_size':500, 'width':7}
nx.draw_shell(Gpetersengraph,**options,nlist=[range(6,11),range(6)], with_labels = True, font_weight='bold')
plt.show()
```

## Barbell graph : 


#### A Barbell graph is a special type of undirected graph consisting of two non overlapping n-vertex CLIQUES together with a single edge that has an endpoint in each clique 

### Create a barbell graph 

```python
GBarbell = nx.barbell_graph(5,1)

print('Nodes of barbell graph : ')
print(GBarbell.nodes())
print('Edges of barbell graph :')
print(GBarbell.edges())

#draw the barbell graph
nx.draw(GBarbell, with_labels=True)
plt.show()
```

## A famous use of graphs : *Communities* , know how to mesure the communities structure 

#### NetworkX includes a wide range of algorithms for various tasks related to network analysis (check https://networkx.org/documentation/stable/reference/algorithms/index.html), among them: the community detection algorithms, which are used to identify groups of nodes that are more densely connected to each other than to the rest of the network 

```python
#import the community module 
from networkx.algorithms import community 
#use the community module 
print('Communities of the graph :')

#generate a community 
communities_generator = community.girvan_newman(GBarbell)

top_level_communities = next(communities_generator)

next_level_communities = next(communities_generator)

sorted(map(sorted, next_level_communities))
```
